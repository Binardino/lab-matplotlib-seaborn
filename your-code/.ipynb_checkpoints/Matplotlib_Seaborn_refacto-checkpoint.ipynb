#### Plot (x, y) and (x, z).
There are 2 ways of doing this. Do in both ways.

**Hint**: Check out the nrows, ncols and index arguments of subplots. 

Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
#option 1 : creating both plots with straightforward plt.plot option

#creating the figure framework
plt.figure(figsize=(20,10))
#creating two subplots with plt.subplot
plt.subplot(1,2,1)
plt.plot(x,y, linewidth=5, color='green', linestyle='dashed')

plt.subplot(1,2,2)
plt.plot(x,z, linewidth = 3, color='purple', linestyle=':')
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# using plt.subplots to plot both charts side by side
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,5))

# fig.suptitle() for overall title
fig.suptitle('f(x) = 2x & f(x) = x²',fontsize= 15, fontweight="bold")

#ax[i].set_title() for subplot titles
ax[0].plot(x,y, linewidth=5, color='green', linestyle='dashed')
ax[0].set_title('f(x) = 2x',fontsize= 7)
ax[1].plot(x,z,linewidth = 3, color='purple', linestyle=':')
ax[1].set_title('f(x) = x²',fontsize= 7)
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
y = x**2

z = np.exp(x)
```

```python
fig, ax = plt.subplots(figsize=(20,10))
fig.suptitle('f(x)=x² & f(x)=exp(x)', fontsize=20, fontweight='bold')
plt.plot(x, y, linewidth=5, color='b', linestyle='dashed')
plt.plot(x, z, linewidth=5, color='r',linestyle='dashdot')

ax.set_xscale('linear')
ax.set_yscale('log')
```

#### As a bonus challenge, try to add a legend to the plot.

```python
fig, ax = plt.subplots(figsize=(20,10))
fig.suptitle('f(x)=x² & f(x)=exp(x)', fontsize=20, fontweight='bold')
plt.plot(x, y, linewidth=5, color='b', linestyle='dashed', label='f(x)=x²')
plt.plot(x, z, linewidth=5, color='r',linestyle='dashdot', label='f(x)=exp(x)')
plt.legend(loc='upper left', fontsize=20)
ax.set_xscale('linear')
ax.set_yscale('log')
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

```python
#Grouping by median amount of steps per Months
df_months = df_fitbit2.groupby('Months_encoded').agg({'Steps':'median'})
```

```python
df_months
```

```python
# df_weekdays = df_fitbit2.groupby('Work_or_Weekend').agg({'Minutes Lightly Active':'sum','Minutes Fairly Active':'sum','Minutes Very Active':'sum'})
```

```python
df_weekdays = df_fitbit2['Steps'].groupby(df_fitbit2['Work_or_Weekend']).median()
```

```python
df_weekdays = df_fitbit2.groupby('Work_or_Weekend').agg({'Steps':'median'})
```

```python
df_weekdays
```

```python
fig, ax = plt.subplots(1,2, figsize=(17,7))
fig.suptitle('evolution of steps per month & during the week',fontsize= 20, fontweight="bold")

#plotting linechart
ax[0].set_title('evolution of median steps per month', fontsize=15)
ax[0].plot(df_months.index, df_months['Steps'], lw=3, linestyle='--')
ax[0].set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
ax[0].set_xlabel('steps in units')
ax[0].set_ylabel('months')
ax[0].grid(True)

#plotting barchart
ax[1].bar(df_weekdays.index, df_weekdays['Steps'], edgecolor ='b', alpha=0.5)
ax[1].set_title('Median Nb of Steps Week days vs. Week-end', fontsize=15)
# ax[1].set_xticklabels(['Weekdays', 'Week-end'])
plt.xticks(ticks=range(2), labels=['Weekdays', 'Week-end'], rotation=0)
ax[1].set_xlabel('days')
ax[1].set_ylabel('steps in units')
ax[1].grid(True)

#saving the chart
# plt.savefig('fitbit2_Median Nb of Steps Week days vs. Week-end.svg')
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
list_features = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

colours = ['r', 'g', 'b', 'c', 'm', 'y', 'k']

fig, ax = plt.subplots(1,3,figsize=(30,15))

for i in range(3):
    plt.sca(ax[i])
    plt.scatter(df_fitbit2[list_features[i]], df_fitbit2['Steps'], color = random.choice(colours))
    plt.title(f'{list_features[i]} VS. Steps')
    plt.xlabel(list_features[i])
    plt.ylabel('Steps')
plt.savefig("3 Activity variable vs. Steps.svg")
```

#### How does the bin size affect your plot?

**Conclusion**
- Low number of bins : general observation - fewer noise
- High number of bins : greater precision - but may generate more noise and outliers

#### Use the right plot to visualize the column `Pclass`.

```python
# Graph #1 : split of PClass VS. Age
# using catplot boxen

plt.figure(figsize=(12,6))
sns.catplot(x='Pclass', y='Age', kind="boxen", data=df_titanic, height=7,aspect=3)
plt.title('Boxen Catplot PClass VS. Age')
plt.savefig('boxen_catplot_pclass_age.svg')
```

```python
# Graph #1 : split of PClass per Age with Genger filter
# using violinplot

plt.figure(figsize=(20,10))
sns.violinplot(x= df_titanic['Pclass'], y= df_titanic['Age'],
                hue=df_titanic['Gender'],
                split=True,
                inner="quartile",
                palette=["lightblue", "lightpink"]
              )
plt.title('Violinplot PClass VS. Age Gender filtered')
plt.savefig('violinplot_pclass_age_gender.svg')
```

#### We would like to have in one single plot the summary statistics of the feature `Age`. What kind of plot would you use? Plot it. 

```python
# We already used Barplots before, here we'll use a Seaborn boxenplot

plt.figure(figsize=(20,10))
sns.boxenplot(df_titanic['Age'],  orient="h", palette="Set2")
#adding axvline to clearly split the distribution over the median
plt.axvline(df_titanic['Age'].median(), linestyle='dashed', linewidth=5, color='b')
```

#### What does the last plot tell you about the feature `Age`?

The boxenplot displays the several Quartile of the Age distribution :
- the min & max values
- the median (highlighted by the blue dashed axvline)
- the Q1 (25%) & Q3 (75%) quartiles

**Conclusion**
- overall young population : median of 27 years old
- 50% of the population between 23 & 37 years old
- min age = 0 i.e. there were several babies & younglings among the passengers
- max age = 80
